# **SOLUCIÓN EJERCICIO 3**

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)

```python
#Importar las librerías que serán usadas
import numpy as np
import pandas as pd
```

```python
nba_df = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
nba_df.head()
```

```python
nba_df.shape
```

```python
nba_df.info()
```

```python
#Se eliminan las instancias con registros faltantes
nba_df=nba_df.dropna()
nba_df.info()
```

```python
#Se crea un DataFrame para almacenar la suma de los salarios por equipo
team_salaries_df=pd.DataFrame(nba_df.groupby('Team').Salary.sum().sort_values(ascending=False))
team_salaries_df
```

```python
#Se guarda en un dataframe los 5 equipos de la NBA con mayores salarios
top5_salary_teams_df=team_salaries_df.iloc[:5]
print("A continuación se presentan los 5 equipos que destinan mayores valores en los salarios:\n",top5_salary_teams_df)
```

2. Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición

```python
#Se generan 3 grupos de la variable Age para categorizarla
nba_df['Age_Group']=pd.cut(nba_df.Age,3)
nba_df
```

```python
#Se imprime el dataframe agrupado por Age_Group y Posición, realizando el conteo de jugadores según dicha agrupación.
#Posteriormente se genera un Plot con la misma información.
print("A continuación se presenta la cantidad de jugadores por grupo de edad en cada posición: \n",
      nba_df.groupby(['Age_Group','Position']).Name.count().unstack())
nba_df.groupby(['Age_Group','Position']).Name.count().unstack().plot(kind='bar')
```

3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios

```python
nba_df.set_index('Team')
nba_df
#Se crea un dataframe con la información de solo los 5 equipos con mayores salarios.
nba_top_Salary_teams_df=nba_df[nba_df['Team'].isin(top5_salary_teams_df.index.values.tolist())]
nba_top_Salary_teams_df
```

```python
print("A continuación se muestra un pivot table con el salario promedio según la posición para los 5 equipos con mayores salarios:")
#se genera la pivot table con el salario promedio por equipo según cada posición y se imprime este resultado
print(nba_top_Salary_teams_df.pivot_table('Salary',
                                          index='Team',
                                          columns='Position',
                                          aggfunc='mean'))
```
